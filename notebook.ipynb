```python
def maxx(array2d):
    s = []
    for i in range(0,len(array2d)):
        if array2d[i][0]>array2d[i][1]:
            s.append(array2d[i][0])
        else:
            s.append(array2d[i][1])
    return s
```

```python
def rmse(X,Y):
    s = 0
    for i in range(0,len(X)):
        s = s+np.square(X[i]-Y[i])
    s = s/len(X)
    return np.sqrt(s)
```
